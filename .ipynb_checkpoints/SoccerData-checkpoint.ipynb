## Player Dataset
In this project, we want to visualise the data for the soccer players of the 2021/22 Premier League season. The data is collected from: https://www.kaggle.com/datasets/omkargowda/football-players-stats-premier-league-20212022. Some of the columns names are updated for better clarification.  From the csv table, we have 30 columns which describe different things for an individual player. The columns are explained below.

* Player: Name of the player

* Team: The team that the players play for

* Nation: The country that the players play for

* Pos: The postion of the player in the field

* Age: The age of the player

* MP: The number of matches that the player has played

* Starts: The number of the matches that the player has started 

* Min: The number of minutes that the player has played

* 90s: This is the number of minutes played divided by 90

* Gls: This is the number of goals scored by the player

* Ast: This is the number of assists given by the player

* G-PK: This is the number of non penalty goals scored by the player

* PK: This is the number of penalties socred by the player 

* PKatt : This is the number of penalty kicks attempted 

* CrdY : This is the number of yellow cards obtained

* CrdR : This is the number of red cards obtained

* Gls p90 : This is the number of goals scored per 90 mins

* Ast p90: This is the number of assits per 90 mins
 
* G+A p90: This is the number of goals and assists per 90 mins

* npG p90: This is the number of non penalty goals scored per 90 mins

* npG+A p90: This is the number of non penalty goals and assists per 90 mins

* xG : This is the calculation of expected goals 

* npxG : This is the calculation of Non-Penalty Expected Goals

* xA : This is the calculation of Expected Assits

* npxG+xA : This is the calculation of Non-Penalty Expected Goals plus Expected Assists

* xGp90 : This is the calculation fo Expected Goals per 90 mins

* npxGp90 : This is the calculation of Non-Penalty Expected Goals per 90 mins

* xAp90 : This is the calculation of Expected Assits per 90 mins

* npxG+xAp90 : This is the calculation Non-Penalty Expected Goals plus Expected Assists per 90 mins


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('playerdata.csv', encoding='windows-1252')
df.head()
```

```python
df.info()
```

```python
df.hist(bins=10 ,figsize=(16,12), color = 'Green')
plt.show()
```

```python
players = ['Bukayo Saka', 'Thomas Partey']

goals = [11, 2]
xG = [9.7, 2.5]

indices = np.arange(len(players))
bar_width = 0.35

fig, ax = plt.subplots()
ax.bar(indices - bar_width/2, goals, label='Goals', color='blue', width=bar_width, alpha=0.7)
ax.bar(indices + bar_width/2, xG, label='xG(Expected Goals)', color='red', width=bar_width, alpha=0.7)

ax.set_ylabel('Goals and xG(Expected Goals)')
ax.set_title('Comparison of Goals and xG(Expected Goals) for Players')
ax.set_xticks(indices)
ax.set_xticklabels(players)
ax.legend()

plt.show()
```

### From the plot above, we can see that Bukayo Saka outperforms his xG. He has more goals scored than what was expected from him. He was more efficient in his goalscoring. On the other hand, Thomas Partey was underwhelming in his goalscoring. His xG was higher than his goals scored. So, we can conclude that he was inefficient with his finishing. 

```python
players = ['Bruno Fernandes', 'Paul Pogba']

assists = [6, 9]
xA = [8.4, 3.5]

indices = np.arange(len(players))
bar_width = 0.35

fig, ax = plt.subplots()
ax.bar(indices - bar_width/2, assists, label='Assists', color='blue', width=bar_width, alpha=0.7)
ax.bar(indices + bar_width/2, xA, label='xA', color='red', width=bar_width, alpha=0.7)

ax.set_ylabel('Assists and xA')
ax.set_title('Comparison of Assists and xA for Players')
ax.set_xticks(indices)
ax.set_xticklabels(players)
ax.legend()

plt.show()
```

### From the plot above, we can see that Paul Pogba has outperformed his expected assists by a big margin. Therefore, we can say that his teammates were very clinical in converting his passes into goals. On the contrary, Bruno Fernandes was unlucky with his passes as his xA was over 8 but he only got 6 assists. His teammates struggled to convert his passes.  

## Distribution of Player Ages
### Visualize the distribution of player ages to understand the age profile of players  the Premier League.


```python
sns.histplot(df['Age'], bins=10, color='skyblue')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

### From the histogram, we can see that most of players' ages are from 18-34. A few players have ages less than 18 and greater than 34 

```python
top_scorers = df.sort_values(by='Gls', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Gls', y='Player', data=top_scorers)
plt.title('Top 10 Goal Scorers')
plt.xlabel('Goals')
plt.ylabel('Player')
plt.show()
```

### From the barplot, we can see that Mohamed Salah and Son Heung-min were the top scorers of the premier league in the 2021/22 season. They both had 23 goals. The closest to them was Cristiano Ronaldo with 18 goals. 

```python
top_assisters = df.sort_values(by='Ast', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Ast', y='Player', data=top_assisters)
plt.title('Top 10 Goal Assisters')
plt.xlabel('Assists')
plt.ylabel('Player')
plt.show()
```

### From the barplot above, we can see that Mohamed Salah again has the highest number of assists with 13 assists closely followed by his Liverpool teammate Trent Alexander-Arnold with 12 assist. In fact, the joint 3rd top assister is Andrew Robertson who also plays for Liverpool. Looking at this, we can tell that the Liverpool team scored a lot of goal that season.

```python
total_goals = df['Gls'].sum()
total_assists = df['Ast'].sum()

plt.figure(figsize=(6,4))
sns.barplot(x=['Goals', 'Assists'], y=[total_goals, total_assists], palette='coolwarm')
plt.title('Total Goals vs. Assists in the League')
plt.ylabel('Count')
plt.show()
```

```python
team_performance = df.groupby('Team')[['npG p90', 'Ast p90']].mean().reset_index()

plt.figure(figsize=(12,8))
sns.barplot(x='npG p90', y='Team', data=team_performance.sort_values('npG p90', ascending=False), color='lightgreen', label='npG')
sns.barplot(x='Ast p90', y='Team', data=team_performance.sort_values('npG p90', ascending=False), color='lightblue', label='Ast')
plt.title('Average Non-Penalty Goals and Assists by Team')
plt.xlabel('Average Count')
plt.ylabel('Team')
plt.legend()
plt.show()
```

```python
# Comparison of Yellow and Red Cards
players = ['Bukayo Saka', 'Thomas Partey']
yellow_cards = [4, 8]  # Example data, replace with actual data
red_cards = [0, 1]  # Example data, replace with actual data

# Array of indices for the players
indices = np.arange(len(players))

# Plotting the data
fig, ax = plt.subplots()
ax.bar(indices, yellow_cards, label='Yellow Cards', color='yellow', alpha=0.7)
ax.bar(indices, red_cards, bottom=yellow_cards, label='Red Cards', color='red', alpha=0.7)

# Adding labels and title
ax.set_ylabel('Number of Cards')
ax.set_title('Comparison of Yellow and Red Cards for Players')
ax.set_xticks(indices)
ax.set_xticklabels(players)
ax.legend()

# Display the plot
plt.show()
```

```python
# Goal Contribution Ratio
players = ['Bruno Fernandes', 'Paul Pogba']
goals = [12, 6]  # Example data, replace with actual data
assists = [8, 9]  # Example data, replace with actual data
total_goals = [goals[i] + assists[i] for i in range(len(players))]

# Calculate contribution ratio
goal_ratio = [goals[i] / total_goals[i] for i in range(len(players))]
assist_ratio = [assists[i] / total_goals[i] for i in range(len(players))]

# Array of indices for the players
indices = np.arange(len(players))

# Plotting the data
fig, ax = plt.subplots()
ax.bar(indices, goal_ratio, label='Goals', color='blue', alpha=0.7)
ax.bar(indices, assist_ratio, bottom=goal_ratio, label='Assists', color='green', alpha=0.7)

# Adding labels and title
ax.set_ylabel('Contribution Ratio')
ax.set_title('Goal Contribution Ratio for Players')
ax.set_xticks(indices)
ax.set_xticklabels(players)
ax.legend()

# Display the plot
plt.show()
```
